```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn import datasets
import numpy as np
import pandas as pd
from pyswarm import pso
```

```python
# Load dataset
lung_data = pd.read_csv('/content/cancer patient data sets.csv')
```

```python
lung_data.head()
```

```python
print(lung_data.duplicated().sum())
```

```python
lung_data.drop(["index", "Patient Id"], axis = 1, inplace = True, errors='ignore')
print(lung_data)
```

```python
x=lung_data.drop(['Level'],axis=1)
y=lung_data['Level']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state=42)
```

```python
# Standardize the features
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
# Define the machine learning model
model = RandomForestClassifier()

# Train the model with all features
model.fit(x_train, y_train)
```

```python
# Predict the test set results
y_pred = model.predict(x_test)
```

```python
accuracy_before = accuracy_score(y_test, y_pred)
precision_before = precision_score(y_test, y_pred, average='weighted')
recall_before = recall_score(y_test, y_pred, average='weighted')
f1_before = f1_score(y_test, y_pred, average='weighted')
```

```python
# Define PSO objective function
def objective_function(selected_features, x_train, x_test, y_train, y_test):
    selected_features_indices = np.where(selected_features == 1)[0]
    if len(selected_features_indices) == 0:
        return 1e6  # Penalty for empty feature set
    model = RandomForestClassifier()
    model.fit(x_train[:, selected_features_indices], y_train)
    y_pred = model.predict(x_test[:, selected_features_indices])
    return -accuracy_score(y_test, y_pred)  # Negative accuracy for minimization

# Define bounds for PSO optimization
n_features = x_train.shape[1]
lb = [0] * n_features  # Lower bound for feature selection (binary)
ub = [1] * n_features  # Upper bound for feature selection (binary)

# Apply Particle Swarm Optimization for feature selection
selected_features, _ = pso(objective_function, lb, ub, args=(x_train, x_test, y_train, y_test))

selected_features_indices = np.where(selected_features == 1)[0]
print("Selected features:", selected_features_indices)

# Train the model with selected features
model.fit(x_train[:, selected_features_indices], y_train)

# Predict the test set results
y_pred = model.predict(x_test[:, selected_features_indices])

# Evaluate model performance (after feature selection)
accuracy_after = accuracy_score(y_test, y_pred)
precision_after = precision_score(y_test, y_pred, average='weighted')
recall_after = recall_score(y_test, y_pred, average='weighted')
f1_after = f1_score(y_test, y_pred, average='weighted')

print("\nPerformance After Feature Selection:")
print(f"Accuracy: {accuracy_after}")
print(f"Precision: {precision_after}")
print(f"Recall: {recall_after}")
print(f"F1 Score: {f1_after}")

# Compare performance improvement
accuracy_diff = accuracy_after - accuracy_before
precision_diff = precision_after - precision_before
recall_diff = recall_after - recall_before
f1_diff = f1_after - f1_before

print("\nPerformance Improvement After Feature Selection:")
print(f"Accuracy Improvement: {accuracy_diff}")
print(f"Precision Improvement: {precision_diff}")
print(f"Recall Improvement: {recall_diff}")
print(f"F1 Score Improvement: {f1_diff}")
```
